```python
import numpy as np
x=np.array(range(1,6))
y=np.array(range(10,15))
def gradiant_descent(x,y):
    m=b=0
    iter=10
    n=len(x)
    lr=0.044
    for j in range(iter): 
        y_pred=m*x+b
        cost=1/n*np.sum((y-y_pred)**2)
        md=-2/n*np.sum(x*(y-y_pred))
        bd=-2/n*np.sum(y-y_pred)
        m=m-lr*md
        b=b-lr*bd 
        print(f"m={m}    b={b}      cost={cost}")    
    return m,b
best_fit_line=gradiant_descent(x,y)
best_fit_line      
```
