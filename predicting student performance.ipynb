```python
col_iqr=['age','G1','G2','absences']
for i in col_iqr:
    q1=df[i].quantile(0.25)
    q3=df[i].quantile(0.75)
    iqr=q3-q1
    lb=q1-1.5*iqr
    ub=q3+1.5*iqr
    df=df[(df[i]>=lb)&(df[i]<=ub)]
```

```python
import numpy as np
for i in col_iqr:
    df[i]=np.log1p(df[i])
```

```python
df.drop(columns=['freetime','famrel','Dalc','Walc'],axis=1)
```

```python
x=df.drop(columns='G3',axis=1)
y=df.G3
y.unique()
y=df['G3'].apply(lambda x:1  if x>=10 else 0)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split,KFold,cross_val_score,GridSearchCV
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import RobustScaler,OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import f1_score,confusion_matrix,classification_report
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```python
step_cat=[('impute',SimpleImputer(strategy='most_frequent')),
          ('encode',OneHotEncoder(handle_unknown='ignore',drop='first'))
          ]
pipe_cat=Pipeline(steps=step_cat)


step_num=[('impute',SimpleImputer(strategy='median')),
          ('scale',RobustScaler())
          ]
pipe_num=Pipeline(steps=step_num)


transform=[('num',pipe_num,col_iqr),
           ('cat',pipe_cat,cat_col)]
col_t=ColumnTransformer(transformers=transform)
pipe=Pipeline(steps=[('preprocess',col_t),
                     ('model',LogisticRegression())])
```

```python
kf=KFold(n_splits=5,shuffle=True,random_state=42)
param=[{'model':[LogisticRegression()],
        'model__C':np.arange(0.0001,1,0.1)},
        {'model':[KNeighborsClassifier()],
         'model__n_neighbors':np.arange(1,10)},
         {'model':[SVC()],
          'model__kernel':['rbf','linear', 'poly', 'sigmoid'],
          'model__degree':np.arange(1,10),
          'model__gamma':['scale','auto'],
          'model__coef0':np.linspace(0.1,2,10)},
          {'model':[RandomForestClassifier()],
           'model__criterion':['gini', 'entropy', 'log_loss'],
           'model__max_depth':np.arange(1,20),
           'model__n_estimators':np.arange(1,10)},
           {'model':[DecisionTreeClassifier()],
            'model__max_depth':np.arange(1,10),
            'model__splitter': ['best', 'random'] },
            {'model':[GaussianNB()]},
             ]
```

```python
grid=GridSearchCV(pipe,param_grid=param,cv=kf,scoring='f1',n_jobs=-1)
grid.fit(x_train,y_train)
y_pred=grid.predict(x_test)
grid.best_score_
```

```python
results = pd.DataFrame(grid.cv_results_)

col = results[['param_model', 'mean_test_score', 'rank_test_score']].copy()

col['param_model'] = col['param_model'].astype(str)

results_sorted = col.sort_values(by='mean_test_score', ascending=False)

best = col.groupby('param_model')['mean_test_score'].mean().sort_values(ascending=False)
```

```python
plt.figure(figsize=(12,6))
plt.bar(best.index,best.values,width=0.5)
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
grid.best_estimator_
```

```python
cross_v=cross_val_score(grid.best_estimator_,x,y,cv=kf,scoring='f1')
print(cross_v)
print(cross_v.mean())
```

```python
con_mat=confusion_matrix(y_test,y_pred)
sns.heatmap(con_mat,annot=True,cmap='coolwarm')
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
sp = ConfusionMatrixDisplay(con_mat)
sp.plot()
plt.show()
```

```python
report=classification_report(y_test,y_pred)
print(report)
```
